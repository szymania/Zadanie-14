importuję moduły

```python
import matplotlib.pyplot as plt
import numpy as np
```

opisuję zmienne

```python
f = 1
A = 1
t = np.linspace(0, 1, 256)
sygnal = A*np.sin(2*np.pi*f*t)
```

tworzę wykres

```python
plt.ylabel('Wartość funkcji [-]')
plt.xlabel('Czas [s]')
plt.title('Amplituda')
plt.plot(t, sygnal, label = 'A = 1')
```
